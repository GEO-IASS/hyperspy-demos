## EDS-TEM quantification of buried nanostructure

Using a machine learning technique, such as independent component analysis (ICA), the composition of buried nanostructure can be accurately measured as it has been demonstrated by D. Roussow et al. (Nano Letters, In Press, 2015). Using the same data, this notebook reproduces the main result of this article.

## Representative spectrum from bare cores

To obtain a representative spectrum of the bare nanoparticles, the low intensity of Pt L${\alpha}$ is masked

```python
pt_la = c.get_lines_intensity(['Pt_La'])[0]
```

```python
mask = pt_la > 6
```

```python
axes = utils.plot.plot_images([mask, pt_la * mask], axes_decor=None, colorbar=None,
                       label=['Mask', 'Pt L${\\alpha}$ intensity'])
```

The sum over the particles is used as a bare core spectrum

```python
s_bare = (c * mask).sum(0).sum(0)
```

## Comparison and quantification

Stack the spectrum of bare particles and the first ICA component.

```python
s_bare.change_dtype('float')
s = utils.stack([s_bare, s_bss], new_axis_name='Bare or BSS')
s.metadata.General.title = 'Bare or BSS'
```

```python
axes = utils.plot.plot_spectra(s, style='mosaic', legend=['Bare particles', 'BSS #0'])
```

X-ray intensities measurement with background subtraction

```python
w = s.estimate_background_windows()
```

```python
s.plot(background_windows=w, navigator='slider')
```

Refinement of the windows position

```python
w
```

```python
w[1, 0] = 8.44
w[1, 1] = 8.65
```

```python
s.plot(background_windows=w, navigator='slider')
```

```python
sI = s.get_lines_intensity(background_windows=w)
```

Set up the kfactors for Fe K${\alpha}$ and Pt L${\alpha}$

```python
#From Brucker software (Esprit)
kfactors = [1.450226, 5.075602]
```

Quantify with Cliff Lorimer

```python
composition = s.quantification(intensities=sI, kfactors=kfactors,
                 plot_result=True, composition_units='atomic')
```

## Going further

### Isolating the nanoparticles

Further image processing with skimage. Apply a watershed transformation to isolate the nanoparticles.

```python
from scipy.ndimage import distance_transform_edt, label
from skimage.morphology import watershed
from skimage.feature import peak_local_max
```

Adapted from http://scikit-image.org/docs/dev/auto_examples/plot_watershed.html
- Transform the mask into a distance map
- Find local maxima
- Apply the watershed to the distance map using the local maximum as seed (markers)

```python
distance = distance_transform_edt(mask.data)
local_maxi = peak_local_max(distance, indices=False,
                            min_distance=2, labels=mask.data)
labels = watershed(-distance, markers=label(local_maxi)[0],
                   mask=mask.data)
```
