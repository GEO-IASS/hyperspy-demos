## Exctract core from bare cores

The low intensity of Pt L${\alpha}$ is masked

```python
pt_la = c.get_lines_intensity(['Pt_La'])[0]
```

```python
mask = pt_la > 6
```

```python
utils.plot.plot_images([mask, pt_la * mask], axes_decor='off',
                       label=['Mask', 'Pt L${\\alpha}$ intensity'])
```

The sum over the particles is used as a bare core spectrum

```python
s_bare = (c * mask).mean(0).mean(0)
```

## Going further

### Image processing: Watershed

Further image processing with skimage. Apply a watershed transformation to isolated the nanoparticles.

```python
from scipy import ndimage
from skimage import restoration
from skimage.morphology import watershed
from skimage.feature import peak_local_max
```

Adapted from http://scikit-image.org/docs/dev/auto_examples/plot_watershed.html
- Transform the mask into a distance map
- Find the local maximum
- Apply the watershed to the distance map using the local maximum as seed

```python
distance = ndimage.distance_transform_edt(mask.data)
local_maxi = peak_local_max(distance, indices=False,
                            min_distance=2, labels=mask.data)
labels = watershed(-distance, markers=ndimage.label(local_maxi)[0],
                   mask=mask.data)
```

```python
utils.plot.plot_images(
    [pt_la, mask, signals.Image(distance), signals.Image(labels)],
    axes_decor='off', per_row=2,
    label=['Pt L${\\alpha}$ intensity', 'Mask', 'Distances', 'Separated particles'])
```
