# Hyperspy Tutorial

## Visualisation of 4D EDS/SEM mapping

Hyperspy can be easily used on data with more dimension than the classical EDS mapping. EDS/SEM can be extended to three-dimensional microanalysis with a "slice-and-view" approach. From such data, a 3D stack of X-ray line intensity maps can be extracted, de-noised and plotted in 3D, as shown in this notebook with data from a Ni-based supperalloy.

### Energy axis scaling and microscope parameters

Energy axis scaling and microscope parameters can be found in the one pixel spectrum, exported as ".msa"

```python
s1p = load('Ni_superalloy_1pix.msa')
s1p.set_signal_type('EDS_SEM')
```

```python
s.metadata
```

### Spatial axes scaling

This parameters have to be found in the acquisition software and set either with the gui either by directly accessing the `axes_manager`

```python
units = '${\mu}m$'
scaling = {'x': 0.05, 'y': 0.05, 'z': 0.1}
```

```python
s.axes_manager.gui()
```

```python
for i, ax in enumerate(scaling.keys()):
    s.axes_manager[i].name = ax
    s.axes_manager[ax].units = units
    s.axes_manager[ax].scale = scaling[ax]    
```

```python
s.plot()
```

## Going further

### Scaling Mayavi isosurface 

Generating a scalar field allows a better control, such as the scaling

```python
mlab.figure(bgcolor=(1, 1, 1))
#Create a scalar field and scale it
src = mlab.pipeline.scalar_field(ni.data)
src.name = 'Ni intensity'
src.spacing = [ni.axes_manager[ax].scale for ax in ['z', 'x', 'y']]
#Plot the isosurface
mlab.pipeline.iso_surface(src, contours=[85], color=(1., 0.5, 0.2)) 
mlab.outline(color=(0, 0, 0))
```

```python
#Display in the notebook
imshow(mlab.screenshot()[1:])
```

### De-noising the stack of images

The following python packages can be used to de-noise the Ni L${\alpha}$ intensity maps.

```python
from skimage.restoration import denoise_tv_chambolle
from scipy.ndimage import median_filter
```

For the filters to be applied on all 3 dimension, the signal dimension is set to 3

```python
ni = s.get_lines_intensity(['Ni_La'])[0]
ni.axes_manager.set_signal_dimension(3)
ni.change_dtype('float')
```

A list of copy is created and the filters are applied

```python
nis = [ni.deepcopy() for i in range(3)]
nis[1].map(denoise_tv_chambolle, n_iter_max=4)
nis[2].map(median_filter, size=3)
```

For the dislpay, the signals are set as stack of image

```python
for i, n in enumerate(nis):
    nis[i] = n.as_image([0, 1])
    nis[i].metadata.General.title = 'Ni L${\\alpha}$ intensity' + \
        ['Raw', 'TV', 'median'][i]
utils.plot.plot_images(nis, axes_decor=None)
```
